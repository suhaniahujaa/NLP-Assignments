```python
import pandas as pd
import random

# --- Template-Based Data Generation ---
# Define templates and variables for each intent to create diverse queries.
generation_templates = {
    "order_problem": {
        "templates": [
            "My order is {issue}.",
            "I have a problem with order {order_id}.",
            "There's something wrong with my delivery.",
            "My package {status}.",
            "I'm having an issue with my recent purchase.",
            "My order seems to be {issue}."
        ],
        "issue": ["missing", "lost", "not delivered", "stuck in transit", "delayed"],
        "status": ["hasn't shown up", "never arrived", "is very late"],
        "order_id": ["#12345", "#98765", "#54321"]
    },
    "track_order": {
        "templates": [
            "Where is my order {order_id}?",
            "Can you track my package?",
            "What's the status of my shipment?",
            "Give me an update on my delivery.",
            "Find order {order_id}.",
            "I need tracking information for my purchase."
        ],
        "order_id": ["#ABC-123", "#XYZ-987", "#FGH-456"]
    },
    "return_item": {
         "templates": [
            "I want to return {item}.",
            "How do I start a return process?",
            "What's the procedure for sending back {item}?",
            "I need to send this back.",
            "Can I get a refund for {item}?"
        ],
        "item": ["this product", "my purchase", "the shoes", "the laptop"]
    },
    "cancel_subscription": {
        "templates": [
            "Cancel my {subscription_type} subscription.",
            "How do I stop my membership?",
            "I want to end my subscription.",
            "Please terminate my {subscription_type} plan.",
            "Unsubscribe me from the service."
        ],
        "subscription_type": ["monthly", "yearly", "premium", "gold"]
    },
    "password_reset": {
        "templates": [
            "I forgot my password.",
            "How to reset my password?",
            "I can't log in, need password help.",
            "Send me a password reset link.",
            "My password isn't working."
        ]
    },
    "refund_policy": {
        "templates": [
            "What is your refund policy?",
            "How do refunds work?",
            "Tell me about your money-back guarantee.",
            "Can I get my money back if I'm not satisfied?"
        ]
         },
    "technical_support": {
        "templates": [
            "I need tech support for {product}.",
            "The {feature} isn't working.",
            "My device is frozen.",
            "Help with a technical issue regarding {product}.",
            "The app is crashing."
        ],
        "product": ["my camera", "the app", "the software", "my account"],
        "feature": ["login button", "checkout process", "search bar"]
    },
    "contact_support": {
        "templates": [
            "How can I contact customer service?",
            "I need to speak to a person.",
            "What's your support phone number?",
            "Is there a live chat option?",
            "Give me your contact details."
        ]
    },
    "product_damage": {
        "templates": [
            "My {item} arrived damaged.",
            "The product is broken.",
            "I received a defective item.",
            "The {item} is cracked.",
            "There's a scratch on my new {item}."
        ],
        "item": ["package", "order", "phone screen", "table"]
         },
    "billing_question": {
        "templates": [
            "I have a question about my bill.",
            "There's a charge I don't recognize on my invoice.",
            "Can you explain this charge?",
            "My payment was declined.",
            "I was overcharged for my last order."
        ]
    },
    "update_shipping": {
        "templates": [
            "I need to update my shipping address.",
            "Can I change the delivery address for order {order_id}?",
            "My address is wrong on my profile.",
            "How to change my shipping info?"
        ],
        "order_id": ["#555-_111", "#222_333"]
    },
    "change_payment": {
        "templates": [
            "How do I change my payment method?",
            "I need to update my credit card information.",
            "Can I use a different card for my subscription?",
            "Update my billing details."
        ]
    },
    "leave_feedback": {
        "templates": [
            "I want to leave some feedback.",
            "Where can I write a review?",
            "I have a suggestion for your service.",
            "My feedback on the recent experience."
        ]
    },
    "report_bug": {
        "templates": [
            "I think I found a bug on your website.",
             "The mobile app has a glitch.",
            "Something is not working correctly on the checkout page.",
            "I want to report an error."
        ]
    },
    "operating_hours": {
        "templates": [
            "What are your operating hours?",
            "When are you open?",
            "What are your support hours?",
            "Are you available on weekends?"
        ]
    },
    "apply_discount": {
        "templates": [
            "How do I apply a discount code?",
            "My promo code isn't working.",
            "Where do I enter the coupon?",
            "Can I use this discount on my order?"
        ]
    },
    "login_troubleshoot": {
        "templates": [
            "I'm having trouble logging in.",
            "I can't sign into my account.",
            "My login is not working.",
            "The website won't let me log in."
        ]
    },
    "wrong_item": {
        "templates": [
            "I received the wrong item.",
            "You sent me the incorrect product.",
            "This is not what I ordered.",
            "I got {wrong_item} instead of {correct_item}."
        ],
        "wrong_item": ["a red shirt", "the small size", "the wrong model"],
        "correct_item": ["a blue one", "the large size", "the pro version"]
    },
    "use_giftcard": {
        "templates": [
            "How do I use a gift card?",
            "Where to enter my gift card code?",
            "My gift card balance isn't working.",
            "Can I pay with a gift card?"
        ]
    },
    "request_demo": {
        "templates": [
            "I'd like to request a demo of your product.",
            "Can I see a demonstration of the software?",
            "How to schedule a product demo?",
            "I'm interested in a demo."
        ]
    },
    "greeting": {
        "templates": [
            "hello", "hi", "hey there", "good morning", "greetings",
            "hello bot", "good afternoon", "what's up", "yo"
        ]
    }
}

# --- Generation Logic ---
generated_data = []
intents = list(generation_templates.keys())

# Generate around 60 examples for each intent
for _ in range(60):
    for intent in intents:
        template_data = generation_templates[intent]
        template = random.choice(template_data["templates"])

        # Fill in the placeholders in the template
        query = template
        if "{issue}" in query:
            query = query.replace("{issue}", random.choice(template_data["issue"]))
        if "{status}" in query:
            query = query.replace("{status}", random.choice(template_data["status"]))
        if "{order_id}" in query:
            query = query.replace("{order_id}", random.choice(template_data["order_id"]))
        if "{item}" in query:
            query = query.replace("{item}", random.choice(template_data["item"]))
        if "{subscription_type}" in query:
            query = query.replace("{subscription_type}", random.choice(template_data["subscription_type"]))
        if "{product}" in query:
            query = query.replace("{product}", random.choice(template_data["product"]))
        if "{feature}" in query:
            query = query.replace("{feature}", random.choice(template_data["feature"]))
        if "{wrong_item}" in query:
            query = query.replace("{wrong_item}", random.choice(template_data["wrong_item"]))
        if "{correct_item}" in query:
            query = query.replace("{correct_item}", random.choice(template_data["correct_item"]))

        generated_data.append({'query': query, 'intent': intent})

# Create the final DataFrame
new_df = pd.DataFrame(generated_data)

# --- Display Results ---
print(f"Successfully generated {len(new_df)} new training examples.")
print("\nSample of the generated data:")
print(new_df.head())
print("...")
print(new_df.tail())

print("\nDistribution of generated intents:")
print(new_df['intent'].value_counts())
```
